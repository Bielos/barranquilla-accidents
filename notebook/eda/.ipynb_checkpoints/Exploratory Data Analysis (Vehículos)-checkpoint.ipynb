# Análisis Introductorio de Datos: Búsqueda de patrones en Vehículos
Objetivos:
* Establecer los patrones generales relacionados con los vehículos involucrados en accidentes de tránsito en la ciudad de Barranquilla desde 2015 hasta 2018.

## Lectura de Datos para Análisis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

vehiculos_df = pd.read_csv('../../data/processed/Vehiculos_clean.csv')
vehiculos_df.head(10)
```

## Principales Aspectos a Describir
Este dataset es el más pequeño de los 3, con las variables disponibles se pretende dar la siguiente estadística descriptiva:
* Cantidad de vehículos involucrados en accidentes en cada año.
* Cantidad de vehículos accidentados por clase de vehículo y año.
* Cantidad de vehículos accidentados por tipo de servicio y año.
* Cantidad de vehículos accidentados por clase de accidente y año.

### Cantidad de Vehículos

```python
vehicles_per_year_df = vehiculos_df.groupby("AÑO_ACCIDENTE").agg({"CANTIDAD_VEHICULOS":sum})
vehicles_per_year_df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.barplot(x=vehicles_per_year_df.index.values, y=np.transpose(vehicles_per_year_df.values)[0] , ax=ax)
ax.set(title="Cantidad de Vehñiculos Accidentados de 2015 a 2018")
plt.show()
```

De la gráfica anterior se puede ver que la cantidad de vehiculos totales accidentados por año ronda los 12,000 vehículos.

```python
vehicles_per_year_df.describe()
```

En promedio hay 12,467 vehículos involucrados en accidentes cada año desde 2015 hasta 2018 con una desviación estándar de 546 vehículos.

### Cantidad de Vehículos por Clase de Vehículos

```python
vehicles_per_type_df = vehiculos_df.groupby(["CLASE_VEHICULO_ACCIDENTADO","AÑO_ACCIDENTE"], as_index=False).agg({"CANTIDAD_VEHICULOS":sum})
vehicles_per_type_df.head(10)
```

```python
%matplotlib inline
clase_vehiculos = vehiculos_df['CLASE_VEHICULO_ACCIDENTADO'].unique()
fig = plt.figure(figsize=(8, 70))
for i,clase in enumerate(clase_vehiculos):
    ax = fig.add_subplot(len(clase_vehiculos),1,i+1)
    data = vehicles_per_type_df[vehicles_per_type_df['CLASE_VEHICULO_ACCIDENTADO']==clase]
    sns.barplot(x='CLASE_VEHICULO_ACCIDENTADO', y='CANTIDAD_VEHICULOS', hue='AÑO_ACCIDENTE', data=data, ax=ax)
    ax.set(xlabel=None, ylabel='Cantidad de Vehículos')
plt.tight_layout()
plt.show()
```

Los automoviles, con una cantidad de vehículos entre los 5,000 y 6,000 anual, son el tipo de vehículo que más accidentes y su tendencia parece mantenerse constante desde el 2015 hasta el 2018, año donde hubo la mayor reducción.

El resto de vehículos presentan un comportamiento centrado en la media de su cantidad de vehículos accidentados por año, el único caso que llama la atención es el de la bicicletas, cuya tendencia viene en amuento (explicada, quizás, por un aumento en el uso de las mismas en las grandes ciudades por efecto de las campañas que promueven su uso).

### Cantidad de Vehículos por Tipo de Servicio

```python
vehicles_per_service_df = vehiculos_df.groupby(["SERVICIO_VEHICULO_ACCIDENTADO","AÑO_ACCIDENTE"], as_index=False).agg({"CANTIDAD_VEHICULOS":sum})
vehicles_per_service_df
```

```python
servicio_vehiculos = vehiculos_df['SERVICIO_VEHICULO_ACCIDENTADO'].unique()
fig = plt.figure(figsize=(8, 20))
for i,servicio in enumerate(servicio_vehiculos):
    ax = fig.add_subplot(len(servicio_vehiculos),1,i+1)
    data = vehicles_per_service_df[vehicles_per_service_df['SERVICIO_VEHICULO_ACCIDENTADO']==servicio]
    sns.barplot(x='SERVICIO_VEHICULO_ACCIDENTADO', y='CANTIDAD_VEHICULOS', hue='AÑO_ACCIDENTE', data=data, ax=ax)
    ax.set(xlabel=None, ylabel='Cantidad de Vehículos')
plt.tight_layout()
plt.show()
```

Los vehículos de tipo particular son los que más accidentes reportan, con una media de 7,505 vehículos accidentados por año por una tendencia centrada en la media. En segundo lugar, los vehículos públicostambién presentan una cantidad de accidentes promedio anual centrada en la media, la cual es de 4,735 vehículos accidentados.

Adicionalmente, los vehículos de tipo diplomático presentan una tendencia de disminución, los oficiales tienen una tendencia centrada en la media y los desconocidos vienen en aumento.

### Cantidad de Vehículos por Clase de Accidente

```python
vehicles_per_accident_df = vehiculos_df.groupby(["CLASE_ACCIDENTE","AÑO_ACCIDENTE"], as_index=False).agg({"CANTIDAD_VEHICULOS":sum})
vehicles_per_accident_df
```

```python
clase_accidente = vehiculos_df['CLASE_ACCIDENTE'].unique()
fig = plt.figure(figsize=(8, 20))
for i,accidente in enumerate(clase_accidente):
    ax = fig.add_subplot(len(clase_accidente),1,i+1)
    data = vehicles_per_accident_df[vehicles_per_accident_df['CLASE_ACCIDENTE']==accidente]
    sns.barplot(x='CLASE_ACCIDENTE', y='CANTIDAD_VEHICULOS', hue='AÑO_ACCIDENTE', data=data, ax=ax)
    ax.set(xlabel=None, ylabel='Cantidad de Vehículos')
plt.tight_layout()
plt.show()
```

El atropello es el tipo de accidente que más vehículos registra desde el 2015 hasta el 2018, en éste ultimo año hubo un incremento de aproximadamente 10.6% en la media de los accidentes, la cual hasta el año 2017 era de 209 vehículos, y luego del incremento en 2018 quedó en 231 vehículos.

Los accidentes por choque tiene un comportamiento centrado en la media de 12,143 vehículos accidentados por año. Los accidentes por caída de ocupante vienen en aumento (explicado, tal vez, por un incremento en la cantidad de motocicletas en la ciudad).
