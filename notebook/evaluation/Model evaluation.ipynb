# Evaluación de los modelos
**Objetivo:** Elaborar y ejecutar pruebas que permitan evaluar los resultados en relación a los criterios de éxito definidos

## Creación del dataset de prueba
Se selecciona todo 2018 como dataset de prueba (365 registros) y se evalúa el modelo en términos de su MSE y su MAE.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../../src/modeling/dataset_final.csv')
df['FECHA_ACCIDENTE'] = pd.to_datetime(df['FECHA_ACCIDENTE'], format="%Y-%m-%d %H:%M:%S")
df.head()
```

## Evaluación del modelo SARIMA
Para la evaluación del modelo SARIMA se utilizaron los datos de 2015 hasta 2017 con la intención de medir la eficiencia de la predicción con el año 2018. El mejor modelo fue SARIMA(0,1,1)(2,1,2,12) a continuación se mide la eficiencia del modelo al predecir el año 2018.

### Cargar modelo desde binarios

```python
import pickle
stepwise_fit = pickle.load(open('../../model/test models/SARIMA_model.sav', 'rb'))
stepwise_fit.summary()
```

### Predicción 2018

```python
test = df.set_index('FECHA_ACCIDENTE').loc['2018-01-01':]
y_pred = stepwise_fit.predict(n_periods=365)

fig = plt.figure()
ax = fig.add_subplot()
sns.lineplot(x=test.index, y=test['CANTIDAD_ACCIDENTES'], ax=ax)
sns.lineplot(x=test.index, y=y_pred, ax=ax)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('MSE: {0}'.format(mean_squared_error(test, y_pred)))
print('MAE: {0}'.format(mean_absolute_error(test, y_pred)))
```

El modelo tiene un ajuste con tendencia general de disminución de los accidentes (línea roja) con un MSE de 29.81 y MAE de 4.47. Esto lo que quiere decir es que el modelo se comporta bien manteniendo la media de los accidentes pero si se penaliza la precisión con el cuadrado de la distancia (MSE) el resultado es alto.

## Evaluación del modelo Prophet
Para la evaluación del modelo Prophet se procede de la misma forma a cargar los binarios del modelo para luego predecir el año 2018 y comparar su rendimiento con el modelo SARIMA.

### Cargar modelo desde binarios

```python
import pickle
custom_model = pickle.load(open('../../model/test models/prophet_model.sav', 'rb'))
print(custom_model)
```

### Predicción 2018

```python
test = test.reset_index()

future = custom_model.make_future_dataframe(periods=365)
forecast = custom_model.predict(future)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.lineplot(x=test['FECHA_ACCIDENTE'], y=test['CANTIDAD_ACCIDENTES'], ax=ax)
sns.lineplot(x=test['FECHA_ACCIDENTE'], y=forecast['yhat'].iloc[-365:].values, ax=ax)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('MSE: {0}'.format(mean_squared_error(test['CANTIDAD_ACCIDENTES'].values, forecast['yhat'].iloc[-365:].values)))
print('MAE: {0}'.format(mean_absolute_error(test['CANTIDAD_ACCIDENTES'].values, forecast['yhat'].iloc[-365:].values)))
```

Con el modelo personalizado de Prophet se puede ver por la gráfica que el fit es mucho mejor que el del modelo SARIMA, el modelo prophet no solo se anticipa a los picos repentinos en la frecuencia de los accidentes sino que las tendencias generales de aumento y disminución son muy bien interpretadas para todo el año.

El MSE de 21.76 indica una notable disminución en la predicción de altos y bajos, mientras que el MAE de 3.73 explica lo bien que el modelo está manteniendose cerca de la predicción en cada día. 

## Conclusión

Se mantiene entonces el modelo Prophet personalizado como el ganador, se entrena entonces el modelo con la totalidad de datos disponible y se guardan sus binarios para la fase de despliegue.

```python
from fbprophet import Prophet
import holidays

df.columns = ['ds', 'y']

custom_model = Prophet(
    growth="linear",
    seasonality_mode="additive",
    daily_seasonality=False,
    weekly_seasonality=False,
    yearly_seasonality=False,
).add_seasonality(
    name='daily',
    period=1,
    fourier_order=10
).add_seasonality(
    name='weekly',
    period=7,
    fourier_order=20
).add_seasonality(
    name='yearly',
    period=365.25,
    fourier_order=10
)
custom_model.add_country_holidays(country_name='CO')
custom_model.fit(df)
```

```python
future = custom_model.make_future_dataframe(periods=365)
forecast = custom_model.predict(future)
fig1 = custom_model.plot(forecast)
```

```python
fig2 = custom_model.plot_components(forecast)
```
